```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def read_df(file_title,criteria):
    try:
        df=pd.read_csv(file_title, sep="," )
        d={"number":int(df.columns[0]),criteria: int(df.columns[1])}
        df1=pd.DataFrame(data=d,index=[0])
        return (df1)
    except:
        return(None)   
```

```python
df_nbA= read_df("nbopAvecMemo/"+str(5)+".csv","nbop")
df_nbS= read_df("nbopSansMemo/"+str(5)+".csv","nbop")
df_TA = read_df("TempsAvecMemo/"+str(5)+".csv","temps")
df_TS = read_df("TempsSansMemo/"+str(5)+".csv","temps")
```

```python
for i in range(6,200):
    df_nbA=df_nbA.append(read_df("nbopAvecMemo/"+str(i)+".csv","nbop"),ignore_index=True)
    df_nbS=df_nbS.append(read_df("nbopSansMemo/"+str(i)+".csv","nbop"),ignore_index=True)
    df_TA=df_TA.append(read_df("TempsAvecMemo/"+str(i)+".csv","temps"),ignore_index=True)
    df_TS=df_TS.append(read_df("TempsSansMemo/"+str(i)+".csv","temps"),ignore_index=True)
```

```python
df_nbS = df_nbS.drop(27)
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(df_nbS.number,df_nbS.nbop, color = 'green')
plt.title('Le nombre d\'appels récursifs')
plt.xlabel('Le nombre généré n')
plt.ylabel('nbop')
plt.legend(['Algo Recursif Simple'])
plt.ylim(0,200000)
plt.savefig('nbS.jpg',dpi=184)
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(df_nbA.number,df_nbA.nbop, color = 'green')
plt.title('Le nombre d\'appels récursifs')
plt.xlabel('Le nombre généré n')
plt.ylabel('nbop')
plt.legend(['Algo Récursif avec Memo'])
#plt.ylim(0,200000)
plt.savefig('nbA.jpg',dpi=184)
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(df_TS.number,df_TS.temps/1000, color = 'green')
plt.title('Le temps dépensé en secondes')
plt.xlabel('Le nombre généré n')
plt.ylabel('Temps en s')
plt.legend(['Algo Récursif'])
#plt.ylim(0,3)
plt.savefig('TA.jpg',dpi=184)
plt.show()
```
